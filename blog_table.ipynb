```python
import urllib2
from bs4 import BeautifulSoup

url = "http://guaminsects.net/anr/aubrey_blog_posts"

page = urllib2.urlopen(url).read()
html = BeautifulSoup(page)
```

```python
table_body = ''
table = html.find('table')
trs = table.findAll('tr')

for tr in trs[1:]: # skip header row
    tds = tr.findAll('td')
    timestamp = tds[0].text.strip()
    if timestamp >= '2014':
        link = tds[1].find('a')
        link_text = link.text
        link_ref = link['href']
        href = r'\href{http://guaminsects.net' + link_ref + r'}{' + link_text + r'}'
        table_body += r'{} & {} \\'.format(timestamp, href) + '\n'
#print(table_body)
```

```python
tex = r'''
\begin{longtable}{p{.2\textwidth} p{.75\textwidth}}
\caption{\label{blog-table}Public blog posts on \emph{guaminsects.net/anr} posted 2014-15}
\tabularnewline
\hline
\textbf{Date} & \textbf{Title} \\
\hline
\endfirsthead
\multicolumn{2}{c}%
{\tablename\ \thetable\ -- \textit{Continued from previous page}} \\
\hline
\textbf{Date} & \textbf{Title} \\
\hline
\endhead
\hline \multicolumn{2}{r}{\textit{Continued on next page}} \\
\endfoot
\hline
\endlastfoot
'''

tex += table_body
tex += r'\end{longtable}'
#print(tex)
f = open("blog-table.tex", "w")
f.write(tex)
f.close()
```
